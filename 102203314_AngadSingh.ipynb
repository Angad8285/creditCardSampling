```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Creditcard_data.csv')
dataset.shape
```

```python
class_count_0, class_count_1 = dataset['Class'].value_counts()
class_count_0, class_count_1 
```

```python
class_0 = dataset[dataset['Class'] == 0]
class_1 = dataset[dataset['Class'] == 1]
class_0.shape, class_1.shape
```

```python
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1]
X
```

```python
y
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state = 42)
x_ros, y_ros = ros.fit_resample(X, y)
```

```python
from collections import Counter
Counter(y), Counter(y_ros)
```

```python
dataset = pd.concat([x_ros.reset_index(drop=True),y_ros.reset_index(drop=True)], axis=1)
dataset
```

```python
size1 = 305
n = dataset.shape[0]
interval = (n // size1)
```

```python
# Sample 1 - Systematic random sampling
import random
random_number = random.randint(0, interval)
def systematic_sampling(df, step):
    indexes = np.arange(random_number, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
```

```python
sample1 = systematic_sampling(dataset, interval)
sample1
```

```python
from sklearn.model_selection import train_test_split
X = sample1.iloc[:, :-1].values
y = sample1.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
```

```python
# Model 1 : Logistic Regression
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Model 2 : SVM
from sklearn.svm import SVC
model2 = SVC()
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# Model 3 : Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier()
model3.fit(X_train, y_train)
```

```python
y_pred = model3.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# Model 4 : Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
model4 = GaussianNB()
model4.fit(X_train, y_train)
```

```python
y_pred = model4.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# Model 5 : K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
model5 = KNeighborsClassifier()
model5.fit(X_train, y_train)
```

```python
y_pred = model5.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# Sample 5 - Bootstrap Sampling
samples = 10
bootstrap_data = [
    dataset.sample(n = 384, replace=True, random_state=i).reset_index(drop=True)
    for i in range(samples)
]
```

```python
# Model 1 - Logistic Regression
accuracies = []
for i, bootstrap_sample in enumerate(bootstrap_data):
    X = bootstrap_sample.drop(columns=["Class"])
    y = bootstrap_sample["Class"]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i, stratify=y)

    model = LogisticRegression(max_iter=1000, random_state=i)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

np.mean(accuracies)
```
